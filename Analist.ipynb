# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
data= pd.read_csv('data_harga.csv',header=3)
data.rename(columns={'Unnamed: 0':'Komoditas'}, inplace=True)
data["Komoditas"] = data["Komoditas"].str.replace(r'^\d+\.\s*','', regex=True)
data.describe()
```

# Exploratory Data Analysis (EDA)




```python
# top 5 Nilai tertinggi Komoditas
max5=data.sort_values(by='Tahunan', ascending=False).head(10)
max5
```

```python
# TOP 5 NILAI TERENDAH KOMODITAS
min5=data.sort_values(by='Tahunan').head()
min5
```

```python
# Nilai rata-rata setiap bulan
data.mean(numeric_only=True)
```

# Data Visualization

```python
# Top 10 Nilai Tahunan Komoditas
top10=data.sort_values(by='Tahunan', ascending=False).head(10)
top10=top10.sort_values(by='Tahunan', ascending=True) # Urutkan kembali untuk visualisasi
top10
plt.figure(figsize=(13,10))
xe=plt.barh(top10["Komoditas"], top10["Tahunan"])
for i in range(len(xe)):
    plt.text(xe[i].get_width(), xe[i].get_y() + xe[i].get_height()/2, f'{xe[i].get_width():.2f}', va='center')
plt.title("Top 10 Nilai Tahunan Komoditas")
plt.show()
```

```python
komoditas = data[data["Komoditas"]=="Daging Dan Hasil-Hasilnya"]

komoditas.iloc[:,1:13].T.plot()
plt.title("Tren Harga Daging 2025")
plt.show()
```
